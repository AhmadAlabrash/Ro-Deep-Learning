```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler  
```

```python
df=pd.read_csv('/content/homeprices.csv')
df
```

```python
sa =MinMaxScaler()
x_scaled = sa.fit_transform(df.drop('price',axis=1))
y_scaled = sa.fit_transform(df[['price']])
y_scaled =y_scaled.reshape(y_scaled.shape[0])
```

```python
def gradeint_descint( x , y ,alpha ,epochs):
  total_samples = x.shape[0]
  result =pd.DataFrame( index= ['epoch_number','w','b','loss'])


  w = np.random.random(x.shape[1])
  b = 0

  for i in range(epochs):

      y_pre = np.dot(w,x.T) + b
      

      wd = -(2/total_samples) * (np.dot(x.T , (y - y_pre )))
      bd = -(2/total_samples) * (np.sum(y - y_pre))

      w = w - alpha * wd
      b = b - alpha * bd

      loss = np.mean(np.square(y -y_pre))
      
      if(i %10 == 0):
       print(f" {i} w is {w} , b is : {b} , loss is : {loss}")
       result['epoch_number','w','b','loss']=([i,w,b,loss])
      

  return result
gradeint_descint(x_scaled,y_scaled , 0.01 , 500)
```

```python
def stochastic_gradeint_descint( x , y ,alpha ,epochs):
  total_samples = x.shape[0]
  result =pd.DataFrame( index= ['epoch_number','w','b','loss'])


  w = np.random.random(x.shape[1])
  b = 0

  for i in range(epochs):
      random_sample =np.random.randint(1,total_samples-1)
      x_sample =x[random_sample]
      y_sample = y[random_sample]

      y_pre = np.dot(w,x_sample.T) + b
      

      wd = -(2/total_samples) * (np.dot(x_sample.T , (y_sample - y_pre )))
      bd = -(2/total_samples) * (np.sum(y_sample - y_pre))

      w = w - alpha * wd
      b = b - alpha * bd

      loss = np.mean(np.square(y_sample -y_pre))
      
      if(i %10 == 0):
       print(f" {i} w is {w} , b is : {b} , loss is : {loss}")
       result['epoch_number','w','b','loss']=([i,w,b,loss])
      

  return result
stochastic_gradeint_descint(x_scaled,y_scaled , 0.01 , 500)
```

```python
def mini_batch( x , y ,alpha ,epochs):
  total_samples = x.shape[0]
  result =pd.DataFrame( index= ['epoch_number','w','b','loss'])


  w = np.random.random(x.shape[1])
  b = 0

  for i in range(epochs):
    
      random_sample =np.random.randint(2,total_samples-3)
     
      x_sample =x[random_sample:(random_sample+3)]
      y_sample = y[random_sample:(random_sample+3)]

      y_pre = np.dot(w,x_sample.T) + b
      

      wd = -(2/total_samples) * (np.dot(x_sample.T , (y_sample - y_pre )))
      bd = -(2/total_samples) * (np.sum(y_sample - y_pre))

      w = w - alpha * wd
      b = b - alpha * bd

      loss = np.mean(np.square(y_sample -y_pre))
      
      if(i %10 == 0):
       print(f" {i} w is {w} , b is : {b} , loss is : {loss}")
       result['epoch_number','w','b','loss']=([i,w,b,loss])
      

  return result
mini_batch(x_scaled,y_scaled , 0.01 , 500)
```
